```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('Tests/Scenario 1/Test1/aggregated_data.csv')  # Adjust path if necessary
df['timestamp'] = pd.to_datetime(df['timestamp'])  # Convert 'timestamp' column to datetime if needed

# Convert DataFrame columns to numpy arrays to avoid indexing issues
timestamps = df['timestamp'].values
ground_truth = df['ground_truth'].values
markers = df['markers'].values
ranges = df['range'].values
filtered = df['filtered'].values

# Create figure and axes
fig, axs = plt.subplots(3, 1, figsize=(12, 18), sharex=True)  # 3 rows, 1 column, sharing x axis

# Colors for the plots
ground_truth_color = '#377eb8'  # Blue for ground truth
marker_color = '#ff7f00'       # Orange for markers
range_color = '#4daf4a'        # Green for range
filtered_color = '#e41a1c'     # Red for filtered

# Plot 1: Timestamp vs Ground Truth and Marker
axs[0].plot(timestamps, ground_truth, label='Ground Truth', marker='o', linestyle='-', color=ground_truth_color)
axs[0].plot(timestamps, markers, label='Markers', marker='x', linestyle='--', color=marker_color)
axs[0].set_title('Ground Truth and Markers over Time')
axs[0].set_ylabel('Articulation angle (radians)')
axs[0].legend()
axs[0].grid(True)

# Plot 2: Timestamp vs Ground Truth and Range
axs[1].plot(timestamps, ground_truth, label='Ground Truth', marker='o', linestyle='-', color=ground_truth_color)
axs[1].plot(timestamps, ranges, label='Range', marker='x', linestyle='--', color=range_color)
axs[1].set_title('Ground Truth and Range over Time')
axs[1].set_ylabel('Articulation angle (radians)')
axs[1].legend()
axs[1].grid(True)

# Plot 3: Timestamp vs Ground Truth and Filtered
axs[2].plot(timestamps, ground_truth, label='Ground Truth', marker='o', linestyle='-', color=ground_truth_color)
axs[2].plot(timestamps, filtered, label='Filtered', marker='x', linestyle='--', color=filtered_color)
axs[2].set_title('Ground Truth and Kalman Filter over Time')
axs[2].set_xlabel('Timestamp')
axs[2].set_ylabel('Articulation angle (radians)')
axs[2].legend()
axs[2].grid(True)

plt.tight_layout()
plt.show()

# Save the figure if needed
fig.savefig(f'Tests/Scenario 1/SC1_line_plot.png', format='png')  # Uncomment and use actual directory path
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1.inset_locator import inset_axes

# Sample data
x = range(100)
y = [value ** 2 for value in x]

# Create the main figure and axis
fig, ax = plt.subplots(figsize=(10, 5))

# Plot the main data
ax.plot(x, y, label='Main plot')
ax.set_title('Full Data View with Zoom on a Segment')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')

# Create an inset axis with a specified position and size relative to the parent axis
axins = inset_axes(ax, width="30%", height="30%", loc=2)  # loc=2 is "upper left"

# Plot the same data in the inset but for a smaller range
x1, x2, y1, y2 = 20, 40, 400, 1600  # Define the limits of the zoomed area
axins.plot(x, y, label='Zoom')
axins.set_xlim(x1, x2)
axins.set_ylim(y1, y2)
axins.set_title('Zoom Inset')
axins.grid(True)

# Optionally, connect the inset to the main plot with lines or shading
# This part can be customized or omitted as needed
from mpl_toolkits.axes_grid1.inset_locator import mark_inset
mark_inset(ax, axins, loc1=1, loc2=3, fc="none", ec='0.5')  # Adds lines with corners at loc1 and loc3

plt.show()
```
